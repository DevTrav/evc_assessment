```python
import pandas as pd
from pathlib import Path 
```

```python
# Create dataframes with clean vendor data 
vendor_x_data = pd.read_csv('resources/vendor_x_data.csv', low_memory=False)
vendor_y_data = pd.read_csv('resources/vendor_y_data.csv', low_memory=False)
```

```python
# Merge data and create a new csv

all_vendor_data = pd.concat([vendor_x_data, vendor_y_data], axis=0)

all_vendor_data.sample(10)
```

```python
# Convert data to csv for type comparison across vendor dataframes
filepath = Path('/Users/traverslaville/Desktop/Classwork/interview/evc_assessment/data_wrangling/resources/all_vendor_data.csv')

all_vendor_data.to_csv(filepath)
```
